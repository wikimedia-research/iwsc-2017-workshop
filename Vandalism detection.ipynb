```python
import json
import requests
from sseclient import SSEClient as EventSource

def get_damage_proba(rev_id):
    response = requests.get('https://ores.wikimedia.org/v3/scores/enwiki/{0}/damaging'.format(rev_id))
    try:
        return round(response.json()['enwiki']['scores'][str(rev_id)]['damaging']['score']['probability']['true'], 2)
    except:
        return None

url = 'https://stream.wikimedia.org/v2/stream/recentchange'
for event in EventSource(url):
    if event.event == 'message':
        try:
            change = json.loads(event.data)
        except ValueError:
            pass
        else:
            if 'revision' in change and change['wiki'] == 'enwiki':
                rev_id = change['revision']['new']
                proba = get_damage_proba(rev_id)
                print("https://en.wikipedia.org/?diff={0}".format(rev_id), 
                      proba, "check" if proba and proba > 0.5 else "OK")
```
